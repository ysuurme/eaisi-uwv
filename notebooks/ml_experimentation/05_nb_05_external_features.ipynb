```python
# ============================================================
# MULTIVARIATE ML FORECASTING
# LOAD MULTIPLE CBS TABLES
# ============================================================

# ------------------------------------------------------------
# Imports
# ------------------------------------------------------------

import numpy as np
import pandas as pd

from src.config import DIR_DB_SILVER
from src.utils.m_query_database import f_query_database
from sklearn.ensemble import GradientBoostingRegressor

# ============================================================
# LOAD ABSENTEEISM TABLE
# 80072NED
# ============================================================

query_abs = '''
SELECT *
FROM "80072ned_silver"
'''

df_abs = f_query_database(
    db_path=DIR_DB_SILVER,
    query=query_abs,
    return_type="pandas"
)

print("Absenteeism shape:")
print(df_abs.shape)

# ============================================================
# LOAD LABOR VOLUME TABLE
# 85920NED
# ============================================================

query_labor = '''
SELECT *
FROM "85920ned_silver"
'''

df_labor = f_query_database(
    db_path=DIR_DB_SILVER,
    query=query_labor,
    return_type="pandas"
)

print("\nLabor volume shape:")
print(df_labor.shape)

# ============================================================
# LOAD COMPENSATION TABLE
# 85917NED
# ============================================================

query_comp = '''
SELECT *
FROM "85917ned_silver"
'''

df_comp = f_query_database(
    db_path=DIR_DB_SILVER,
    query=query_comp,
    return_type="pandas"
)

print("\nCompensation shape:")
print(df_comp.shape)

print("\n===== LABOR TABLE COLUMNS =====")
print(df_labor.columns.tolist())

print("\n===== COMPENSATION TABLE COLUMNS =====")
print(df_comp.columns.tolist())

display(df_labor.head())
display(df_comp.head())
```

```python
# ============================================================
# CREATE TEMPORAL FEATURES
# FOR NEW TABLES
# ============================================================

def prepare_periods(df):

    # --------------------------------------------------------
    # Quarterly only
    # --------------------------------------------------------

    df = (
        df[
            df["Perioden"].str.contains("KW")
        ]
        .copy()
    )

    # --------------------------------------------------------
    # Year
    # --------------------------------------------------------

    df["year"] = pd.to_numeric(
        df["Perioden"].str[:4],
        errors="coerce"
    )

    # --------------------------------------------------------
    # Quarter
    # --------------------------------------------------------

    df["quarter"] = (
        df["Perioden"]
        .str.extract(r"KW(\d{2})")[0]
    )

    df["quarter"] = pd.to_numeric(
        df["quarter"],
        errors="coerce"
    )

    # --------------------------------------------------------
    # Quarter end date
    # --------------------------------------------------------

    df["period_enddate"] = (
        pd.PeriodIndex.from_fields(
            year=df["year"].astype(int),
            quarter=df["quarter"].astype(int),
            freq="Q"
        )
        .to_timestamp(how="end")
        .normalize()
    )

    return df


# ============================================================
# APPLY
# ============================================================

df_labor = prepare_periods(df_labor)

df_comp = prepare_periods(df_comp)

print(df_labor.shape)
print(df_comp.shape)

print(
    df_labor[
        "BedrijfstakkenBranchesSBI2008_Title"
    ].nunique()
)

print(
    df_comp[
        "BedrijfstakkenBranchesSBI2008_Title"
    ].nunique()
)
```

```python
# ============================================================
# UNIQUE SECTORS
# ============================================================

abs_sectors = set(
    df_abs[
        "BedrijfskenmerkenSBI2008_Title"
    ].unique()
)

labor_sectors = set(
    df_labor[
        "BedrijfstakkenBranchesSBI2008_Title"
    ].unique()
)

comp_sectors = set(
    df_comp[
        "BedrijfstakkenBranchesSBI2008_Title"
    ].unique()
)

# ============================================================
# COMMON SECTORS
# ============================================================

common_sectors = (
    abs_sectors
    .intersection(labor_sectors)
    .intersection(comp_sectors)
)

print("\nCOMMON SECTORS:")
print(sorted(common_sectors))

print("\nCOUNT:")
print(len(common_sectors))
```

```python
# ============================================================
# PREPARE ABSENTEEISM TABLE
# ============================================================

# ------------------------------------------------------------
# Keep quarterly observations only
# ------------------------------------------------------------

df_abs = (
    df_abs[
        df_abs["Perioden"].str.contains("KW")
    ]
    .copy()
)

# ------------------------------------------------------------
# Year
# ------------------------------------------------------------

df_abs["year"] = pd.to_numeric(
    df_abs["Perioden"].str[:4],
    errors="coerce"
)

# ------------------------------------------------------------
# Quarter
# ------------------------------------------------------------

df_abs["quarter"] = (
    df_abs["Perioden"]
    .str.extract(r"KW(\d{2})")[0]
)

df_abs["quarter"] = pd.to_numeric(
    df_abs["quarter"],
    errors="coerce"
)

# ------------------------------------------------------------
# Quarter end date
# ------------------------------------------------------------

df_abs["period_enddate"] = (
    pd.PeriodIndex.from_fields(
        year=df_abs["year"].astype(int),
        quarter=df_abs["quarter"].astype(int),
        freq="Q"
    )
    .to_timestamp(how="end")
    .normalize()
)

print(df_abs.shape)
```

```python
# ============================================================
# KEEP ONLY COMMON SECTORS
# ============================================================

common_sectors = list(common_sectors)

# ------------------------------------------------------------
# ABSENTEEISM
# ------------------------------------------------------------

df_abs_filtered = (
    df_abs[
        df_abs[
            "BedrijfskenmerkenSBI2008_Title"
        ].isin(common_sectors)
    ]
    .copy()
)

# ------------------------------------------------------------
# LABOR
# ------------------------------------------------------------

df_labor_filtered = (
    df_labor[
        df_labor[
            "BedrijfstakkenBranchesSBI2008_Title"
        ].isin(common_sectors)
    ]
    .copy()
)

# ------------------------------------------------------------
# COMPENSATION
# ------------------------------------------------------------

df_comp_filtered = (
    df_comp[
        df_comp[
            "BedrijfstakkenBranchesSBI2008_Title"
        ].isin(common_sectors)
    ]
    .copy()
)

print(df_abs_filtered.shape)
print(df_labor_filtered.shape)
print(df_comp_filtered.shape)
```

```python
# ============================================================
# ABSENTEEISM MODEL TABLE
# ============================================================

df_abs_model = df_abs_filtered[
    [
        "period_enddate",
        "year",
        "quarter",
        "BedrijfskenmerkenSBI2008_Title",
        "Ziekteverzuimpercentage_1"
    ]
].copy()

df_abs_model.columns = [
    "period_enddate",
    "year",
    "quarter",
    "sector",
    "absenteeism"
]

display(df_abs_model.head())
```

```python
# ============================================================
# LABOR MODEL TABLE
# ============================================================

# ============================================================
# KEEP ONLY TOTAL WORKFORCE
# ============================================================

df_labor_filtered = (
    df_labor_filtered[
        df_labor_filtered[
            "TypeWerkenden_Title"
        ] == "Totaal"
    ]
    .copy()
)

print(df_labor_filtered.shape)

display(
    df_labor_filtered[
        [
            "TypeWerkenden_Title"
        ]
    ].drop_duplicates()
)
df_labor_model = df_labor_filtered[
    [
        "period_enddate",
        "BedrijfstakkenBranchesSBI2008_Title",
        "WerkzamePersonenSeizoengecorrigeerd_2",
        "BanenSeizoengecorrigeerd_4",
        "GewerkteUrenSeizoengecorrigeerd_6"
    ]
].copy()

df_labor_model.columns = [
    "period_enddate",
    "sector",
    "labor_persons",
    "labor_jobs",
    "labor_hours"
]

display(df_labor_model.head())
```

```python
# ============================================================
# COMPENSATION MODEL TABLE
# ============================================================

df_comp_model = df_comp_filtered[
    [
        "period_enddate",
        "BedrijfstakkenBranchesSBI2008_Title",
        "BeloningSeizoengecorrigeerd_2",
        "LonenSeizoengecorrigeerd_4",
        "Loonkosten_7"
    ]
].copy()

df_comp_model.columns = [
    "period_enddate",
    "sector",
    "compensation",
    "wages",
    "labor_costs"
]

display(df_comp_model.head())
```

```python
# ============================================================
# MERGE ALL TABLES
# ============================================================

df_merged = (
    df_abs_model
    .merge(
        df_labor_model,
        on=[
            "period_enddate",
            "sector"
        ],
        how="left"
    )
    .merge(
        df_comp_model,
        on=[
            "period_enddate",
            "sector"
        ],
        how="left"
    )
)

# ------------------------------------------------------------
# Sort
# ------------------------------------------------------------

df_merged = (
    df_merged
    .sort_values(
        [
            "sector",
            "period_enddate"
        ]
    )
    .reset_index(drop=True)
)

print(df_merged.shape)

display(df_merged.head())

# ============================================================
# VALIDATION
# ============================================================

print("\nMissing Values:\n")

print(
    df_merged.isna().sum()
)

print("\nDate Range:")

print(
    df_merged["period_enddate"].min()
)

print(
    df_merged["period_enddate"].max()
)

print("\nNumber of sectors:")

print(
    df_merged["sector"].nunique()
)
```

```python
# ============================================================
# MISSING ABSENTEEISM BY SECTOR
# ============================================================

missing_abs = (
    df_merged
    .groupby("sector")["absenteeism"]
    .apply(lambda x: x.isna().sum())
    .reset_index()
)

display(missing_abs)

# ============================================================
# MISSING COMPENSATION BY SECTOR
# ============================================================

missing_comp = (
    df_merged
    .groupby("sector")[
        [
            "compensation",
            "wages",
            "labor_costs"
        ]
    ]
    .apply(lambda x: x.isna().sum())
)

display(missing_comp)
```

```python
# ============================================================
# KEEP 2010+
# ============================================================

df_merged = (
    df_merged[
        df_merged["year"] >= 2010
    ]
    .copy()
)

# ============================================================
# DROP MISSING TARGET
# ============================================================

df_merged = (
    df_merged[
        df_merged["absenteeism"].notna()
    ]
    .copy()
)

# ============================================================
# SORT
# ============================================================

df_merged = (
    df_merged
    .sort_values(
        [
            "sector",
            "period_enddate"
        ]
    )
    .reset_index(drop=True)
)

print(df_merged.shape)

print("\nDate Range:")

print(
    df_merged["period_enddate"].min()
)

print(
    df_merged["period_enddate"].max()
)

print("\nMissing Values:\n")

print(
    df_merged.isna().sum()
)
```

```python
# ============================================================
# FORWARD FILL ECONOMIC FEATURES
# ============================================================

economic_cols = [
    "compensation",
    "wages",
    "labor_costs"
]

df_merged[economic_cols] = (
    df_merged
    .groupby("sector")[economic_cols]
    .ffill()
)

# ============================================================
# VALIDATION
# ============================================================

print(
    df_merged[economic_cols]
    .isna()
    .sum()
)
```

```python
# ============================================================
# FEATURE ENGINEERING
# ============================================================

df_features = df_merged.copy()

# ============================================================
# ABSENTEEISM LAGS
# ============================================================

for lag in [1, 2, 3, 4]:

    df_features[f"lag_{lag}"] = (
        df_features
        .groupby("sector")["absenteeism"]
        .shift(lag)
    )

# ============================================================
# ROLLING FEATURES
# ============================================================

df_features["rolling_mean_4"] = (
    df_features
    .groupby("sector")["absenteeism"]
    .transform(
        lambda x:
        x.shift(1).rolling(4).mean()
    )
)

df_features["rolling_std_4"] = (
    df_features
    .groupby("sector")["absenteeism"]
    .transform(
        lambda x:
        x.shift(1).rolling(4).std()
    )
)

# ============================================================
# ECONOMIC LAG FEATURES
# ============================================================

economic_vars = [
    "labor_persons",
    "labor_jobs",
    "labor_hours",
    "compensation",
    "wages",
    "labor_costs"
]

for col in economic_vars:

    df_features[f"{col}_lag_1"] = (
        df_features
        .groupby("sector")[col]
        .shift(1)
    )

# ============================================================
# TIME INDEX
# ============================================================

df_features["time_index"] = np.arange(
    len(df_features)
)

# ============================================================
# TARGET
# NEXT QUARTER ABSENTEEISM
# ============================================================

df_features["target"] = (
    df_features
    .groupby("sector")["absenteeism"]
    .shift(-1)
)

# ============================================================
# DROP MISSING ROWS
# ============================================================

df_features = (
    df_features
    .dropna()
    .reset_index(drop=True)
)

print(df_features.shape)

display(df_features.head())
```

```python
# ============================================================
# ONE-HOT ENCODE SECTORS
# ============================================================

df_model = pd.get_dummies(
    df_features,
    columns=["sector"],
    drop_first=True
)

print(df_model.shape)

display(df_model.head())
```

```python
# ============================================================
# BASELINE FEATURES
# ============================================================

baseline_features = [
    "quarter",
    "time_index",
    "lag_1",
    "lag_2",
    "lag_3",
    "lag_4",
    "rolling_mean_4",
    "rolling_std_4"
]

# ============================================================
# EXTENDED FEATURES
# ============================================================

economic_features = [
    "labor_persons_lag_1",
    "labor_jobs_lag_1",
    "labor_hours_lag_1",
    "compensation_lag_1",
    "wages_lag_1",
    "labor_costs_lag_1"
]

# ------------------------------------------------------------
# Sector dummy columns
# ------------------------------------------------------------

sector_dummies = [
    col
    for col in df_model.columns
    if col.startswith("sector_")
]

# ------------------------------------------------------------
# Final feature list
# ------------------------------------------------------------

extended_features = (
    baseline_features
    + economic_features
    + sector_dummies
)

print(len(extended_features))

# ============================================================
# TARGET
# ============================================================

TARGET = "target"
```

```python
# ============================================================
# IMPORTS
# ============================================================

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error
)

# ============================================================
# BASELINE FEATURES
# ============================================================

baseline_features = [
    "quarter",
    "time_index",
    "lag_1",
    "lag_2",
    "lag_3",
    "lag_4",
    "rolling_mean_4",
    "rolling_std_4"
]

# ============================================================
# TARGET
# ============================================================

TARGET = "target"

# ============================================================
# TIME SERIES CV
# ============================================================

tscv = TimeSeriesSplit(
    n_splits=5
)

# ============================================================
# STORAGE
# ============================================================

results_baseline = []

# ============================================================
# FORCE NUMERIC TARGET
# ============================================================

df_model["target"] = pd.to_numeric(
    df_model["target"],
    errors="coerce"
)

df_model["absenteeism"] = pd.to_numeric(
    df_model["absenteeism"],
    errors="coerce"
)

print(df_model[["target", "absenteeism"]].dtypes)

# ============================================================
# CV LOOP
# ============================================================

for fold, (train_idx, test_idx) in enumerate(
    tscv.split(df_model),
    start=1
):

    print(f"\nFold {fold}")

    # --------------------------------------------------------
    # SPLIT
    # --------------------------------------------------------

    train_fold = df_model.iloc[train_idx]

    test_fold = df_model.iloc[test_idx]

    # --------------------------------------------------------
    # FEATURES
    # --------------------------------------------------------

    X_train = train_fold[baseline_features]

    y_train = train_fold[TARGET]

    X_test = test_fold[baseline_features]

    y_test = test_fold[TARGET]

    # --------------------------------------------------------
    # MODEL
    # --------------------------------------------------------

    model_rf = RandomForestRegressor(
        n_estimators=300,
        max_depth=8,
        min_samples_leaf=5,
        random_state=42,
        n_jobs=-1
    )

    # --------------------------------------------------------
    # TRAIN
    # --------------------------------------------------------

    model_rf.fit(
        X_train,
        y_train
    )

    # --------------------------------------------------------
    # PREDICT
    # --------------------------------------------------------

    pred = model_rf.predict(X_test)

    # --------------------------------------------------------
    # METRICS
    # --------------------------------------------------------

    mae = mean_absolute_error(
        y_test,
        pred
    )

    rmse = np.sqrt(
        mean_squared_error(
            y_test,
            pred
        )
    )

    mape = np.mean(
        np.abs(
            (y_test - pred)
            / y_test
        )
    ) * 100

    # --------------------------------------------------------
    # STORE
    # --------------------------------------------------------

    results_baseline.append({
        "fold": fold,
        "MAE": mae,
        "RMSE": rmse,
        "MAPE": mape
    })

# ============================================================
# RESULTS
# ============================================================

results_baseline = pd.DataFrame(
    results_baseline
).round(3)

display(results_baseline)

# ============================================================
# AVERAGE
# ============================================================

summary_baseline = (
    results_baseline[
        [
            "MAE",
            "RMSE",
            "MAPE"
        ]
    ]
    .mean()
    .to_frame()
    .T
)

summary_baseline.insert(
    0,
    "Model",
    "RF Baseline"
)

summary_baseline = summary_baseline.round(3)

print("\nAVERAGE PERFORMANCE")

display(summary_baseline)
```

```python
# ============================================================
# EXTENDED FEATURES
# ============================================================

economic_features = [
    "labor_persons_lag_1",
    "labor_jobs_lag_1",
    "labor_hours_lag_1",
    "compensation_lag_1",
    "wages_lag_1",
    "labor_costs_lag_1"
]

sector_dummies = [
    col
    for col in df_model.columns
    if col.startswith("sector_")
]

extended_features = (
    baseline_features
    + economic_features
    + sector_dummies
)

print(len(extended_features))

# ============================================================
# STORAGE
# ============================================================

results_extended = []

# ============================================================
# CV LOOP
# ============================================================

for fold, (train_idx, test_idx) in enumerate(
    tscv.split(df_model),
    start=1
):

    print(f"\nFold {fold}")

    # --------------------------------------------------------
    # SPLIT
    # --------------------------------------------------------

    train_fold = df_model.iloc[train_idx]

    test_fold = df_model.iloc[test_idx]

    # --------------------------------------------------------
    # FEATURES
    # --------------------------------------------------------

    X_train = train_fold[extended_features]

    y_train = train_fold[TARGET]

    X_test = test_fold[extended_features]

    y_test = test_fold[TARGET]

    # --------------------------------------------------------
    # MODEL
    # --------------------------------------------------------

    model_rf_ext = RandomForestRegressor(
        n_estimators=300,
        max_depth=8,
        min_samples_leaf=5,
        random_state=42,
        n_jobs=-1
    )

    # --------------------------------------------------------
    # TRAIN
    # --------------------------------------------------------

    model_rf_ext.fit(
        X_train,
        y_train
    )

    # --------------------------------------------------------
    # PREDICT
    # --------------------------------------------------------

    pred = model_rf_ext.predict(X_test)

    # --------------------------------------------------------
    # METRICS
    # --------------------------------------------------------

    mae = mean_absolute_error(
        y_test,
        pred
    )

    rmse = np.sqrt(
        mean_squared_error(
            y_test,
            pred
        )
    )

    mape = np.mean(
        np.abs(
            (y_test - pred)
            / y_test
        )
    ) * 100

    # --------------------------------------------------------
    # STORE
    # --------------------------------------------------------

    results_extended.append({
        "fold": fold,
        "MAE": mae,
        "RMSE": rmse,
        "MAPE": mape
    })

# ============================================================
# RESULTS
# ============================================================

results_extended = pd.DataFrame(
    results_extended
).round(3)

display(results_extended)

# ============================================================
# AVERAGE
# ============================================================

summary_extended = (
    results_extended[
        [
            "MAE",
            "RMSE",
            "MAPE"
        ]
    ]
    .mean()
    .to_frame()
    .T
)

summary_extended.insert(
    0,
    "Model",
    "RF Extended"
)

summary_extended = summary_extended.round(3)

print("\nAVERAGE PERFORMANCE")

display(summary_extended)
```

```python
# ============================================================
# COPY
# ============================================================

df_growth = df_merged.copy()

# ============================================================
# VARIABLES
# ============================================================

economic_vars = [
    "labor_persons",
    "labor_jobs",
    "labor_hours",
    "compensation",
    "wages",
    "labor_costs"
]

# ============================================================
# QUARTER-OVER-QUARTER GROWTH
# ============================================================

for col in economic_vars:

    df_growth[f"{col}_qoq_growth"] = (
        df_growth
        .groupby("sector")[col]
        .pct_change(1)
    )

# ============================================================
# YEAR-OVER-YEAR GROWTH
# ============================================================

for col in economic_vars:

    df_growth[f"{col}_yoy_growth"] = (
        df_growth
        .groupby("sector")[col]
        .pct_change(4)
    )

# ============================================================
# ABSENTEEISM FEATURES
# ============================================================

for lag in [1, 2, 3, 4]:

    df_growth[f"lag_{lag}"] = (
        df_growth
        .groupby("sector")["absenteeism"]
        .shift(lag)
    )

# ============================================================
# ROLLING FEATURES
# ============================================================

df_growth["rolling_mean_4"] = (
    df_growth
    .groupby("sector")["absenteeism"]
    .transform(
        lambda x:
        x.shift(1).rolling(4).mean()
    )
)

df_growth["rolling_std_4"] = (
    df_growth
    .groupby("sector")["absenteeism"]
    .transform(
        lambda x:
        x.shift(1).rolling(4).std()
    )
)

# ============================================================
# TIME INDEX
# ============================================================

df_growth["time_index"] = np.arange(
    len(df_growth)
)

# ============================================================
# TARGET
# ============================================================

df_growth["target"] = (
    df_growth
    .groupby("sector")["absenteeism"]
    .shift(-1)
)

# ============================================================
# DROP NA
# ============================================================

df_growth = (
    df_growth
    .dropna()
    .reset_index(drop=True)
)

print(df_growth.shape)

display(df_growth.head())
```

```python
# ============================================================
# ONE-HOT ENCODE SECTORS
# ============================================================

df_growth_model = pd.get_dummies(
    df_growth,
    columns=["sector"],
    drop_first=True
)

print(df_growth_model.shape)

# ============================================================
# BASE FEATURES
# ============================================================

base_features = [
    "quarter",
    "time_index",
    "lag_1",
    "lag_2",
    "lag_3",
    "lag_4",
    "rolling_mean_4",
    "rolling_std_4"
]

# ============================================================
# GROWTH FEATURES
# ============================================================

growth_features = [
    col
    for col in df_growth_model.columns
    if (
        "qoq_growth" in col
        or
        "yoy_growth" in col
    )
]

# ============================================================
# SECTOR DUMMIES
# ============================================================

sector_features = [
    col
    for col in df_growth_model.columns
    if col.startswith("sector_")
]

# ============================================================
# FINAL FEATURES
# ============================================================

rf_growth_features = (
    base_features
    + growth_features
    + sector_features
)

print(len(rf_growth_features))

# ============================================================
# FORCE TARGET TO NUMERIC
# ============================================================

df_growth_model["target"] = pd.to_numeric(
    df_growth_model["target"],
    errors="coerce"
)

df_growth_model["absenteeism"] = pd.to_numeric(
    df_growth_model["absenteeism"],
    errors="coerce"
)

print(
    df_growth_model[
        ["target", "absenteeism"]
    ].dtypes
)

# ============================================================
# TARGET
# ============================================================

TARGET = "target"

# ============================================================
# STORAGE
# ============================================================

results_growth_rf = []

# ============================================================
# TIME SERIES CV
# ============================================================

tscv = TimeSeriesSplit(
    n_splits=5
)

# ============================================================
# CV LOOP
# ============================================================

for fold, (train_idx, test_idx) in enumerate(
    tscv.split(df_growth_model),
    start=1
):

    print(f"\nFold {fold}")

    # --------------------------------------------------------
    # SPLIT
    # --------------------------------------------------------

    train_fold = df_growth_model.iloc[train_idx]

    test_fold = df_growth_model.iloc[test_idx]

    # --------------------------------------------------------
    # FEATURES
    # --------------------------------------------------------

    X_train = train_fold[rf_growth_features]

    y_train = train_fold[TARGET]

    X_test = test_fold[rf_growth_features]

    y_test = test_fold[TARGET]

    # --------------------------------------------------------
    # MODEL
    # --------------------------------------------------------

    model_growth_rf = RandomForestRegressor(
        n_estimators=300,
        max_depth=8,
        min_samples_leaf=5,
        random_state=42,
        n_jobs=-1
    )

    # --------------------------------------------------------
    # TRAIN
    # --------------------------------------------------------

    model_growth_rf.fit(
        X_train,
        y_train
    )

    # --------------------------------------------------------
    # PREDICT
    # --------------------------------------------------------

    pred = model_growth_rf.predict(X_test)

    # --------------------------------------------------------
    # METRICS
    # --------------------------------------------------------

    mae = mean_absolute_error(
        y_test,
        pred
    )

    rmse = np.sqrt(
        mean_squared_error(
            y_test,
            pred
        )
    )

    mape = np.mean(
        np.abs(
            (y_test - pred)
            / y_test
        )
    ) * 100

    # --------------------------------------------------------
    # STORE
    # --------------------------------------------------------

    results_growth_rf.append({
        "fold": fold,
        "MAE": mae,
        "RMSE": rmse,
        "MAPE": mape
    })

# ============================================================
# RESULTS
# ============================================================

results_growth_rf = pd.DataFrame(
    results_growth_rf
).round(3)

display(results_growth_rf)

# ============================================================
# SUMMARY
# ============================================================

summary_growth_rf = (
    results_growth_rf[
        [
            "MAE",
            "RMSE",
            "MAPE"
        ]
    ]
    .mean()
    .to_frame()
    .T
)

summary_growth_rf.insert(
    0,
    "Model",
    "RF Growth Features"
)

summary_growth_rf = summary_growth_rf.round(3)

print("\nAVERAGE PERFORMANCE")

display(summary_growth_rf)
```
