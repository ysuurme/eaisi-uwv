# Final Forecasting Benchmark & Model Comparison

# Objective

The objective of this notebook is to consolidate and compare all forecasting approaches developed throughout the 05_nb notebook series for predicting quarterly absenteeism rates in the Netherlands.

This notebook serves as:
- the final benchmarking framework,

The notebook compares:

The final goal is to identify the most robust and operational forecasting framework for workforce absenteeism prediction.

The project investigates forecasting Dutch workforce absenteeism using CBS datasets across multiple economic sectors.

- quarterly absenteeism percentage (`Ziekteverzuimpercentage`)

- next-quarter absenteeism prediction

The modeling scope includes:
- national-level forecasting,

- multivariate economic forecasting.

# 3. Forecasting Approaches Developed

The project evolved through multiple forecasting stages.

---

# PHASE 1 — Exploratory Time-Series Analysis

## Objective

Understand:
- trend behavior,
- seasonality,
- volatility,
- structural sector differences.

---

## Key Findings

### Strong quarterly seasonality
Absenteeism consistently showed:
- higher winter absenteeism,
- lower summer absenteeism.

---

### Sector heterogeneity
Different sectors exhibited very different:
- absenteeism levels,
- volatility,
- seasonal patterns.

Examples:
- Healthcare → high absenteeism
- Education → strong yearly cycles
- ICT → more unstable dynamics

---

### Long-term trend evolution
Several sectors displayed:
- gradual structural changes,
- evolving absenteeism baselines.

---

# PHASE 2 — Baseline Forecasting Models

## 2.1 Seasonal Naive Forecast

### Strategy

Forecast next quarter using:
- same quarter from previous year.

Example:
- predict 2025 Q1 using 2024 Q1.

---

## Key Role

This served as:
- the benchmark baseline,
- the minimum acceptable forecasting performance.

---

## Main Insight

Seasonality alone explains a large portion of absenteeism dynamics.

---

# PHASE 3 — Classical Statistical Forecasting

## 3.1 SARIMA Modeling

### Models Used

- SARIMA
- Seasonal differencing
- Quarterly seasonality

---

## Methodology

Models were evaluated using:
- rolling forecasting origin,
- expanding-window validation,
- out-of-sample forecasting.

---

## Key Findings

### Strengths
- captured seasonality well,
- interpretable,
- robust on stable sectors.

### Weaknesses
- struggled with structural shifts,
- limited flexibility,
- weaker on volatile sectors.

# 8. Overall Comparative Findings

# Main Conclusions

## 1. Absenteeism is strongly autoregressive

Historical absenteeism patterns remain the strongest forecasting signal.

Dominant predictors:
- rolling_mean_4
- lag_3

---

## 2. Seasonality is highly important

Quarterly seasonal effects explain a substantial portion of absenteeism variation.

---

## 3. External economic indicators add incremental value

Economic growth indicators improved forecasting modestly.

Most useful transformations:
- QoQ growth
- YoY growth

---

## 4. Global models hide sector heterogeneity

Sector-specific forecasting reveals:
- different dynamics,
- different feature importance,
- different model complexity.

---

## 5. Local forecasting models provide richer insights

Local sector models improved:
- interpretability,
- forecasting specialization,
- economic understanding.

---

## 6. Different sectors require different ML architectures

Examples:
- Education → simpler models
- ICT → deeper nonlinear models
- Healthcare → smoother robust models

# 10. Final Project Conclusion

This project demonstrated that:
- workforce absenteeism can be forecasted effectively using machine learning,
